**ChildClass is inheriting from ParentClass.

ChildClass automatically gains access to all attributes and methods of ParentClass.

Types of Inheritance:

Single Inheritance: One subclass inherits from one superclass.

Multiple Inheritance: One subclass inherits from multiple superclasses.


Multilevel Inheritance: A subclass inherits from a superclass, and another subclass inherits from this subclass.

Hierarchical Inheritance: Multiple subclasses inherit from a single superclass.
Hybrid Inheritance: Combination of two or more types of inheritance.**

```python
class parent:

    def test_parent(self) :
        print("this is my parent class ")
```

```python
class child(parent):
    def test_parent1(self) :
        print("this is my child class ")
```

```python
child_obj = child()
```

```python
child_obj.test_parent()
```

```python
child_obj.test_parent1()
```

```python
class Vehicle:
    def __init__(self, name):
        self.name = name

    def start(self):
        print(f"{self.name} has started.")

    def stop(self):
        print(f"{self.name} has stopped.")


class Car(Vehicle):
    def drive(self):
        print(f"{self.name} is being driven.")


class Motorcycle(Vehicle):
    def ride(self):
        print(f"{self.name} is being ridden.")


car = Car("Tesla Model 3")
motorcycle = Motorcycle("Honda CB500F")

car.start()        # Output: Tesla Model 3 has started.
car.drive()        # Output: Tesla Model 3 is being driven.
car.stop()         # Output: Tesla Model 3 has stopped.

motorcycle.start()  # Output: Honda CB500F has started.
motorcycle.ride()   # Output: Honda CB500F is being ridden.
motorcycle.stop()   # Output: Honda CB500F has stopped.
```
